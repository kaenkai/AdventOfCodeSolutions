```python
class Solution:
    def rotate(self, nums: list[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        nums_ = nums[:]
        for j in range(k):
            nums[j] = nums_[j+n-k]
        for j in range(k, n):
            nums[j] = nums_[j-k]


Solution().rotate([1,2,3,4,5,6,7], 3)
```
